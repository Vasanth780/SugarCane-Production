**Analysis for different Continents**

```python

# Using the "Groupby" feature of Pandas Dataframe to rearrange the dataframe based on groups.

df_continent = df.groupby("Continent").sum()

df_country_concat = pd.DataFrame(df.groupby('Continent')['Country'].agg(', '.join))  # Separating the country names to make them clearly visible

df_continent["Country"] = df_country_concat["Country"]  # Adding the modified country names in the dataframe

df_continent
```

```python

df_continent["number_of_countries"] = df.groupby("Continent")["Country"].count()  #  Creating a new column to count number of countries in each row
```

```python
df_continent
```

**Which continent produces maximum sugarcane?**

```python
df_continent["Production(Tons)"].sort_values(ascending =  False).plot(kind = "bar")  # Bar graph to describe max sugarcane production continent-wise
```

**Do number of countries in a Continent effects production of sugarcane?**

```python
continent_names = df_continent.index.to_list()


sbn.lineplot(data = df_continent,x = "number_of_countries", y= "Production(Tons)" )  # Creating a Line Plot for the groupby'ed data


mat.xticks(df_continent["number_of_countries"], continent_names, rotation =90)


mat.show()
```

**Do the Continent with highest land produces more sugarcane**?

```python

sbn.lineplot(data = df_continent,x = "Acreage(Hectare)", y= "Production(Tons)" )  # Creating a Line Plot for the above analysis
```

**Production distribution by continent**

```python
df_continent["Production(Tons)"].plot(kind = "pie", autopct = "%.2f%%")  # Pie Chart to visualize the distribution of production of Sugarcane across all Conntinents


mat.title('Production Distribution by Continent')


mat.show()
```
